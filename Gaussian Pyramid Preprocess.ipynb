```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.transform import pyramid_gaussian
```

```python
x_train_images = np.load("data/x_train_images.npy")
x_test_images = np.load("data/x_test_images.npy")
```

```python
x_train_images.shape
```

```python
x_test_images.shape
```

```python
img1 = x_train_images[1][0]
print(img1.shape)
```

```python
rows, cols = img1.shape
```

```python
pyramid = tuple(pyramid_gaussian(img1, downscale=2))
```

```python
fig, ax = plt.subplots()
ax.imshow(pyramid[2])
```

```python
x_train_images_100 = np.zeros((3556, 4, 100, 100))
x_train_images_50 = np.zeros((3556, 4, 50, 50))
```

```python
obs, channels, width, height = x_train_images_100.shape
```

```python
for o in range(obs):
    for c in range(channels):
        print("observation:",o, "channel:", c)
        pyramid = pyramid_gaussian(x_train_images[o][c], downscale=2)
        next(pyramid)
        pyramid100 = next(pyramid)
        pyramid50 = next(pyramid)
        x_train_images_100[o][c] = pyramid100
        x_train_images_50[o][c] = pyramid50
```

```python
np.save("x_train_images_100", x_train_images_100)
```

```python
np.save("x_train_images_50", x_train_images_50)
```
